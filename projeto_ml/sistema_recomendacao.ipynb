```python
import pandas as pd

#Indexando pelo id do filme do filme e renomeando colunas
filmes = pd.read_csv("movies.csv")
filmes.columns = ["filmeId", "titulo", "generos"]
filmes = filmes.set_index("filmeId")
filmes.head()
```

```python
#Renomeando colunas
notas = pd.read_csv("ratings.csv")
notas.columns = ["usuarioId", "filmeId", "nota", "momento"]
notas.head()
```

```python
notas.describe()
```

## Primeira tentativa de recomendação.

```python
#Quantidade de votos por filme.
notas["filmeId"].value_counts()
```

```python
filmes.loc[318]
```

```python
total_de_votos = notas["filmeId"].value_counts()
total_de_votos.head()
```

```python
#Criando coluna total de votos
filmes['total_de_votos'] = total_de_votos
filmes.head()
```

```python
#Ordenando de forma decrescente pela quantidade total de votos
filmes.sort_values('total_de_votos', ascending = False).head(10)
```

```python
#Pegando media de votos
nota_media = notas.groupby("filmeId").mean()["nota"]
nota_media.head()
```

```python
#Adicionando coluna de media de votos e ordenando por total de votos
filmes["nota_media"] = nota_media
filmes.sort_values('total_de_votos', ascending = False).head(10)
```

```python
filmes.sort_values('nota_media', ascending = False).head(10)
```

```python
#Ordenando por filmes que tenha mais de 10 votos e pegando a media deles
filmes.query("total_de_votos >= 10").sort_values('nota_media', ascending = False).head(10)
```

```python
#Ordenando por filmes que tenha mais de 50 votos e pegando a media deles
filmes_top_50_votos = filmes.query("total_de_votos >= 50")
filmes_top_50_votos.sort_values('nota_media', ascending = False).head(10)
```

```python
#Localizando filmes, baseado em uma lista de filmes.
eu_assisti = [34, 12, 4, 7, 19, 26, 44]
filmes.loc[eu_assisti]
```

```python
#Filtrando filmes por genero e com media de votos maior que 50, e excluindo os que j[a assiti]
filmes_comedy_romance = filmes_top_50_votos.query("generos == 'Comedy|Romance'")
filmes_comedy_romance.drop(eu_assisti, errors ='ignore').sort_values('nota_media', ascending = False).head(10)
```
